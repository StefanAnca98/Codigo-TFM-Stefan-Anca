```python
import pandas as pd
import os

# Obtener la lista de archivos CSV en el directorio actual
csv_files = [f for f in os.listdir() if f.endswith('.csv')]

# Leer y combinar todos los archivos en un solo DataFrame
dataframes = [pd.read_csv(f, index_col=0) for f in csv_files]

# Concatenar todos los DataFrames
scientific_comments = pd.concat(dataframes)

scientific_comments
```

```python
scientific_comments.to_csv("Full_scientific_comments_dataset.csv", index=False, encoding='utf-8')
```
